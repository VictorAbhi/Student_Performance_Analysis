# Handling missing values

### Finding missing values

```python
df.isnull().sum()
```

## Listing the categorical and the Numerical columns

```python
categorical_columns = df.select_dtypes(include=['category','object']).columns
numerical_columns = df.select_dtypes(include=['int64','float64']).columns
```

### Filling the Missing values in a categorical coluns with mode

```python
df[categorical_columns].isnull().sum()
```

##### knn imputation is more sensitive to outliers

```python
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0]) # multiple variables can be imputed at once
```

```python
df[categorical_columns].isnull().sum()
```

```python
df[numerical_columns].isnull().sum()
```

## Filling numerical columns with mean and median

```python
plt.Figure(figsize=(8, 6))
sns.boxplot(data=df[numerical_columns], orient='h')
plt.xticks(rotation=45)
plt.title('Boxplot of Numerical Columns')
plt.show()
```

```python
# fill numerical_columns with median
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())
```

<p> To determine whether numerical columns follow numerical distribution, you cam perform statistical tests or visual inspections. NOrmal distribution also known as Gaussian Distribution, is charecterized by a bell shaped curve where the data cluster around the mean, with most values close to the mean and fewer values farther away. </p>

<p>Visual inspection: Create visualizations like histogram or QQ plots(Quantile-Quantile plots) to visually assess the distribution of your numerical data. A histogram can show the shape of the data distribution, while a QQ plot compares the quantiles of your data against those expected from a normal distribution. A linear pattern in a QQ plot suggests the data is normally distributed.</p>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Create a figure and axes for subplots
num_cols = len(numerical_columns)
fig, axes = plt.subplots(num_cols, 2, figsize=(16, num_cols * 10))

# Plotting histograms and QQ plots for each numerical column
for i, col in enumerate(numerical_columns):
    # Histogram with KDE
    sns.histplot(df[col], bins=20, kde=True, ax=axes[i, 0])
    axes[i, 0].set_title(f'Histogram of {col}')
    axes[i, 0].set_xlabel('')
    axes[i, 0].set_ylabel('Density')
    axes[i, 0].grid(True, linestyle='--', alpha=0.7)

    # Q-Q plot
    stats.probplot(df[col], dist="norm", plot=axes[i, 1], fit=True)
    axes[i, 1].get_lines()[0].set_markersize(2.0)  # Set size of markers
    axes[i, 1].grid(True, linestyle='--', alpha=0.7)
    axes[i, 1].set_xlabel('Theoretical Quantiles')
    axes[i, 1].set_ylabel('Sample Quantiles')
    axes[i, 1].set_title(f'{col} QQ Plot')

# Adjust layout after all subplots are created
plt.tight_layout(pad=3.0)
plt.show()
```

<p>Statistical Tests: Use statistical methods such as the Shapiro-Wilk test or Kolmogorov-Smirnov test. These tests evaluate whether a sample of data comes from a normal distribution. If the p-value from these tests is greater than a chosen significance level(eg. 0.05), it suggests that the data may follow a normal distribution.</p>

<p>Descriptive Statistics: Calculate descriptive statistics such as mean, median, and standard deviation. For normally distributed data, the mean and median are typically close in value, and about 68% of the data falls within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.</p>
